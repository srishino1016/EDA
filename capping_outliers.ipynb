 CAPPING OUTLIERS 

```python
import pandas as pd
POS=pd.read_csv(r"C:\Users\SrishtiNagu\Desktop\DDRS12\CAPSTONE\DATA\cleaned_data\Super_Final_POS.csv")
```

SALES_D (D:per unit sales)

STEP#1: FINDING QUANTILES

```python


q_l = POS['sales_d'].quantile(0.01) # lower limit of the quantile
q_h = POS['sales_d'].quantile(0.99) # upper limit of the quantile
```

REPLACING MAXIMUM QUANTILES

```python


POS.loc[POS['sales_d']<= q_l, 'sales_d'] = q_l
POS.loc[POS['sales_d']>= q_h, 'sales_d'] = q_h
```

PLOTTING FINAL RESULTS

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
plt.figure(figsize=(16,5))
plt.subplot(1,2,1)
sns.distplot(POS['sales_d'])
plt.show()
```
